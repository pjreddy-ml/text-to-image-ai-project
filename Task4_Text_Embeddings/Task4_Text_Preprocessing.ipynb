# Task 4: Text Preprocessing using Hugging Face Transformers

## Objective

The objective of this task is to preprocess natural language text descriptions using the Hugging Face Transformers library. The text is tokenized and encoded into numerical embeddings, which can be used as inputs for text-to-image generation models such as Stable Diffusion.

## Step 2: Import Required Modules

```python
import torch
from transformers import CLIPTokenizer, CLIPTextModel
```

## Step 3: Load the CLIP Tokenizer and Text Encoder

```python
model_name = "openai/clip-vit-base-patch32"

tokenizer = CLIPTokenizer.from_pretrained(model_name)
text_encoder = CLIPTextModel.from_pretrained(model_name)

print("Model loaded successfully!")
```

## Step 4: Enter a text description

```python
text = "A cute golden retriever playing with a red ball in a green park"

print("Input Text:")
print(text)
```

## Step 5: Generate Text Embeddings

```python
inputs = tokenizer(
    text,
    padding="max_length",
    truncation=True,
    max_length=77,
    return_tensors="pt"
)

print("Input IDs:")
print(inputs["input_ids"])

print("\nAttention Mask:")
print(inputs["attention_mask"])
```

## Step6: Display The Results

```python
with torch.no_grad():
    outputs = text_encoder(**inputs)
    embeddings = outputs.last_hidden_state

print("Embedding Shape:", embeddings.shape)
```

## Step7: Save The Generated Embeddings

```python
import os

os.makedirs("outputs", exist_ok=True)

torch.save(embeddings, "outputs/text_embeddings.pt")

print("Embeddings saved successfully!")
```
