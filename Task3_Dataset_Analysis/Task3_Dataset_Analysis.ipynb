#Import libraries

```python
import torch
import torchvision
from torchvision import datasets, transforms

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from PIL import Image
import os
```

#Download and load the Datasets

```python
from torchvision.datasets import Flowers102

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

train_dataset = Flowers102(
    root="dataset",
    split="train",
    download=True,
    transform=transform
)

val_dataset = Flowers102(
    root="dataset",
    split="val",
    download=True,
    transform=transform
)

test_dataset = Flowers102(
    root="dataset",
    split="test",
    download=True,
    transform=transform
)

print("Training Images :", len(train_dataset))
print("Validation Images :", len(val_dataset))
print("Testing Images :", len(test_dataset))
```

#Explore Dataset Structure

```python
print("Number of Training Images :", len(train_dataset))
print("Number of Validation Images :", len(val_dataset))
print("Number of Testing Images :", len(test_dataset))

total_images = len(train_dataset) + len(val_dataset) + len(test_dataset)

print("\nTotal Images :", total_images)
print("Number of Classes :", 102)
```

#Analyze Image Properties

```python
image, label = train_dataset[0]

print("Image Shape :", image.shape)
print("Image Height :", image.shape[1])
print("Image Width :", image.shape[2])
print("Channels :", image.shape[0])

print("Label :", label)
```

#Dataset Summary

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, ax in enumerate(axes.flat):
    image, label = train_dataset[i]

    ax.imshow(image.permute(1, 2, 0))
    ax.set_title(f"Label: {label}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```
